```python
import skfuzzy as fuzzy
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(0, 10, 500)
y1 = fuzzy.trimf(x, [0, 2.5, 5])
y2 = fuzzy.trimf(x, [2.5, 5, 7.5])
y3 = fuzzy.trimf(x, [5, 7.5, 10])
plt.plot(x, y1, 'b-', x, y2, 'r-', x, y3, 'g-')
plt.grid()
plt.legend([r'$y_1$', r'$y_2$', r'$y_3$'])
```
